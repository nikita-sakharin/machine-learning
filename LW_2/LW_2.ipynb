## Сколько уникальных городов в датафрейме?

```python
df.city_name.nunique() # len(df.city_name.value_counts())
```

```python
df.city_name = df.city_name.astype('category')
df.specialisation = df.specialisation.astype('category')
df.successful_orders = df.successful_orders.astype('float32')
df.fail_orders = df.fail_orders.astype('float32')
df.reorders = df.reorders.astype('float32')
df.info() # about 35% memory saving
```

## Какая специализация имеют самую высокую долю неуспешных заказов (все заказы по конкретному ресторану складываются из успешных, неуспешных заказов и перезаказов)?

```python
groupby = df.groupby('specialisation')[['successful_orders', 'fail_orders', 'reorders']].sum()
groupby
```

```python
groupby['ratio'] = groupby.fail_orders / (groupby.successful_orders + groupby.fail_orders + groupby.reorders)
groupby.ratio
```

```python
groupby.ratio.idxmax()
```

## Какой средний model_score для chain_id 27399 (ответ округлить до 2 знаков после запятой).

```python
score = pd.read_csv('restaurant_scores.csv', sep = ',')
score
```

```python
df_join = df.merge(score, left_on = 'vendor_id', right_on = 'id', how = 'inner')
df_join.drop('id', axis = 1, inplace = True)
df_join
```

```python
df_filter = df_join[df_join.chain_id == 27399]
df_filter
```

```python
round(df_filter.model_score.mean(), 2)
```
